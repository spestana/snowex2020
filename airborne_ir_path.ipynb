```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
filepath = r"C:/Users/steve/OneDrive/Documents/School Stuff/UW/Mountain Hydrology Research Group/MountainGOES/Mountain GOES Paper/Paper 2 - satellite vs ground/Paper 2a - snowex data/data/SNOWEX_LWradiometer_PLANE_2020Feb08_0_5s_smoothed.csv"

radiometer = pd.read_csv(filepath)

radiometer = radiometer[~np.isnan(radiometer.time)]
```

```python
# convert matlab time to python datetime (UTC)
M = radiometer.time.values
base_date = datetime.datetime(1970,1,1,0,0,0)
return_matrix = []
for time in M:
    return_matrix.append(base_date + datetime.timedelta(seconds=time))
times = return_matrix
```

```python
radiometer['datetime'] = times
radiometer.set_index('datetime', inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(15,4))
radiometer.rTemp.plot(ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
radiometer.plot.scatter(x='E', y='N', c='rTemp', cmap='magma', ax=ax)
ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots(figsize=(10,10))
radiometer.plot.scatter(x='N', y='time', c='rTemp', cmap='magma', ax=ax)
```

```python
dt = datetime.timedelta(minutes=2)
dt2 = datetime.timedelta(minutes=10)
imagetime = pd.Timestamp('2020-02-08 195616')
starttime = imagetime - dt
endtime = imagetime + dt

fig, ax = plt.subplots(figsize=(10,10))
radiometer[starttime:endtime].plot.scatter(x='E', y='N', c='rTemp', cmap='magma', alpha=0.5, ax=ax, colorbar=False)
radiometer[imagetime - dt2: imagetime].plot.scatter(x='E', y='N', c='b', s=1, ax=ax, zorder=-999, label='backward direction')
radiometer[imagetime: imagetime + dt2].plot.scatter(x='E', y='N', c='r', s=1, ax=ax, zorder=-999, label='foreward direction')
ax.axvline(218147.5,linestyle='--', color='k')
ax.axvline(239002.5,linestyle='--', color='k')
ax.set_aspect('equal')
plt.legend()
```
