Plot the timeseries of LST and brightness temperatures

```python
fig,ax=plt.subplots(figsize=(15,5))

# specify axes limits
startdate, enddate = pd.Timestamp(2020,2,7),pd.Timestamp(2020,2,10)

# plot land surface temperature
g16_LST_feb.LST.plot(      ax=ax, markeredgecolor='red', markerfacecolor='white', linestyle='none', marker='D', markersize=10, label='GOES-16 Land Surface Temperature', zorder=99)
g16_C07_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 7 Brightness Temperature')
g16_C08_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 8 Brightness Temperature')
g16_C09_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 9 Brightness Temperature')
g16_C10_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 10 Brightness Temperature')
g16_C11_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 11 Brightness Temperature')
g16_C12_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 12 Brightness Temperature')
g16_C13_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 13 Brightness Temperature')
g16_C14_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 14 Brightness Temperature')
g16_C15_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 15 Brightness Temperature')
g16_C16_feb.ref_or_tb.plot(ax=ax, linestyle='none', marker='.', label='GOES-16 Band 16 Brightness Temperature')

# set up x axis
ax.set_xlim(startdate,enddate)
ax.set_xlabel('Date, Time (UTC)')

# set up y axis
ax.set_ylabel('Temperature (K)')

# add legend to the plot
plt.legend(loc=(1.01, 0))

# add a title to the plot
plt.title('GOES ABI Timeseries, Bands 7-16 Brightness Temperature\nGrand Mesa; Feb 2020');
```
